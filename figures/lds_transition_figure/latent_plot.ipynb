```python
import torch
import numpy as np

import matplotlib.pyplot as plt
```

```python
latent_i = 6

means = []
for f in [0.8, 0.9, 1.0, 1.1, 1.2]:
    arr = torch.load(f"decoded_means_{latent_i}_{f}.pt")
    means.append(arr.mean(0)[100:300].cpu().detach().numpy())
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
image = np.hstack(means)
ax.matshow(image.T, cmap="gray")
ax.axis("off")
fig.tight_layout()
plt.show()
```

```python
fig.savefig("manipulationlatent6.pdf", bbox_inches='tight', format="pdf")
```
